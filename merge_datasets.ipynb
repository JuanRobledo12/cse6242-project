```python
import pandas as pd
```

```python
df_moods = pd.read_csv('data_moods.csv')
df_moods_2 = pd.read_csv('data_moods_2.csv')

# Find the 'uri' values in df_moods_2 that match 'id' values in df_moods
matching_uris = df_moods_2['uri'].isin(df_moods['id'])

# Remove these rows from df_moods_2
df_moods_2_cleaned = df_moods_2[~matching_uris]

# Print out the number of rows removed
rows_removed = df_moods_2.shape[0] - df_moods_2_cleaned.shape[0]
print(f"Removed {rows_removed} rows from df_moods_2.")

# Check how many 'id' values in df_moods are also present as 'uri' values in df_moods_2
common_ids = df_moods['id'].isin(df_moods_2_cleaned['uri']).sum()

# Print the number of common values
print(f"There are {common_ids} rows where the 'id' values from df_moods match the 'uri' values from df_moods_2_cleaned.")
```

```python
# Assuming df_moods and df_moods_2_cleaned are already loaded and cleaned
# Rename and reorder columns in df_moods_2_cleaned to match df_moods
df_moods_2_cleaned.rename(columns={
    'uri': 'id',
    'timeSignature': 'time_signature',
    'length': 'length'  # Ensure correct type conversion below
}, inplace=True)

# Assuming merged_df is your DataFrame
# Ensure that the 'length' column is in a timedelta format
df_moods_2_cleaned['length'] = pd.to_timedelta(df_moods_2_cleaned['length'])

# Convert the 'length' column from timedelta to milliseconds
df_moods_2_cleaned['length'] = df_moods_2_cleaned['length'].dt.total_seconds() * 1000  # Convert seconds to milliseconds

# Convert 'length' from object to int in df_moods_2_cleaned
df_moods_2_cleaned['length'] = pd.to_numeric(df_moods_2_cleaned['length'], errors='coerce').fillna(0).astype(int)
```

```python
df_moods_2_cleaned.head()
```

```python


# Define the common columns for merging, based on df_moods
common_columns = [
    'name', 'artist', 'id', 'mood', 'length', 'acousticness',
    'danceability', 'energy', 'instrumentalness', 'liveness',
    'loudness', 'speechiness', 'tempo', 'valence', 'time_signature'
]


# Since df_moods has more columns, we'll align df_moods_2_cleaned with it, adding missing columns with NaNs
for column in df_moods.columns:
    if column not in df_moods_2_cleaned.columns:
        df_moods_2_cleaned[column] = pd.NA  # Add missing columns as NA

# Reorder df_moods_2_cleaned columns to match df_moods
df_moods_2_cleaned = df_moods_2_cleaned[df_moods.columns]

# Merge datasets, ensuring the result follows the format and columns of df_moods
merged_df = pd.concat([df_moods, df_moods_2_cleaned], ignore_index=True)

# After merging, you might want to sort or reset the index if needed
# merged_df = merged_df.sort_values(by=['some_column']).reset_index(drop=True)

merged_df.info()
```

```python
# Assuming merged_df is already created and loaded
# Count the number of duplicated 'id' values
duplicated_ids_count = merged_df['id'].duplicated().sum()  # This counts the number of duplicate occurrences, not including the first occurrence.

# Print the number of duplicated 'id' values
print(f"There are {duplicated_ids_count} duplicated 'id' values in the merged dataset.")
```

```python
# Assuming merged_df is already created and loaded
# Remove rows with duplicated 'id' values, keeping the first occurrence
merged_df = merged_df.drop_duplicates(subset='id', keep='first')

# Recheck the number of duplicated 'id' values after removal
recheck_duplicated_ids_count = merged_df['id'].duplicated().sum()

# Print the number of duplicated 'id' values after removal
print(f"After removal, there are {recheck_duplicated_ids_count} duplicated 'id' values in the merged dataset.")
```

```python
# Finally, save the merged dataset or perform further analysis
merged_df.to_csv('merged_moods_data.csv', index=False)
```
